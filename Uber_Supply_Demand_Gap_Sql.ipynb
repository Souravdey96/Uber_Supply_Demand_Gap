```python
import pandas as pd
import sqlite3
```

## Creating Connection with Sqlite3

```python
df = pd.read_excel('uber_cleaned_data.xlsx')

conn = sqlite3.connect('uber_data.db')

df.to_sql('uber_data', conn, if_exists='replace', index=False)
```

## Querying the Dataset

```python
pd.read_sql("SELECT * FROM uber_data LIMIT 5", conn)
```

## Peak Hour Demand

```python
pd.read_sql('SELECT request_hour, COUNT(*) AS total_requests FROM uber_data GROUP BY request_hour ORDER BY total_requests DESC;', conn)
```

## Airport vs City Demand

```python
pd.read_sql('SELECT pickup_point, COUNT(*) AS total_trips FROM uber_data GROUP BY pickup_point;', conn)
```

## High Value Rides

```python
pd.read_sql('SELECT pickup_point, AVG(trip_duration_min) AS avg_trip_duration FROM uber_data GROUP BY pickup_point;', conn)
```

## Congestion Detection

```python
pd.read_sql('SELECT request_hour, AVG(trip_duration_min) AS avg_trip_time FROM uber_data GROUP BY request_hour ORDER BY avg_trip_time DESC;', conn)
```

## Best Time To Deploy Airport Drivers

```python
pd.read_sql('SELECT time_slot, COUNT(*) AS airport_trips FROM uber_data WHERE pickup_point = "Airport" GROUP BY time_slot ORDER BY airport_trips DESC;', conn)
```

## Best Time To Deploy City Drivers

```python
pd.read_sql('SELECT time_slot, COUNT(*) AS city_trips FROM uber_data WHERE pickup_point = "City" GROUP BY time_slot ORDER BY city_trips DESC;', conn)
```
